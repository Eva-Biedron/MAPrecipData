## 2.3 Selecting Single Stations
### Set up environment

```python
# Import packages
import cartopy
import cartopy.crs as ccrs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load station data.
unique_stations = pd.read_csv(r'C:\Users\15414\Documents\GitHub\MA Precip\MAPrecipData\Data_Products\unique_stations.csv')
#unique_stations.head()
```

```python
# Load precipitation data.
precipitation_flat = pd.read_csv(r'C:\Users\15414\Documents\GitHub\MA Precip\MAPrecipData\Data_Products\precipitation_flat.csv')
#precipitation_flat.head()
```

### Select stations with appropriate records

```python
# Cartopy takes coordinates in within a ([-180°, 180°],[-90°, 90°]) space.
# Data currently requires East/West label to determine directionality (e.g. 72° could be 72°E or 72°W).
# Convert longitude values to negative numbers since they are in the western hemisphere.

unique_stations['Longitude'] = unique_stations['Longitude'] * -1
```

```python
# Select stations whose records end in 2019, are 30+ years long, and have no missing data.
select_stations = unique_stations[(unique_stations['Last_year_collected'] == 2019) &
                                  (unique_stations['First_year_collected'] <= 1988) &
                                  (unique_stations['Months_not_reported'] == 0)
                                 ]
```

```python
# From that, select three stations around Boston metro area.
stations_to_analyze = select_stations[(select_stations['STATION'] == 'COH726') |
                                      (select_stations['STATION'] == 'LYN614') |
                                      (select_stations['STATION'] == 'WEY738')
                                     ]
stations_to_analyze.head()
```

### Graph station locations

```python
ax = plt.axes(projection=ccrs.PlateCarree())

plt.title("Stations with precipitation records longer than 30 years")

# Add coordiantes as axes, but do not show lines over the map itself.
grid_lines = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
grid_lines.top_labels = False
grid_lines.right_labels = False
grid_lines.xlines = False
grid_lines.ylines = False

# Set map extent and show state boundaries.
ax.set_extent([-69.5, -74, 41, 43.5], crs=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.STATES.with_scale('10m'), linewidth=1, facecolor='lightgrey', edgecolor='k')

# Plot stations, stations of interest, and add a legend.
ax.plot('Longitude', 'Latitude', 'o', label='Weather Station', transform=ccrs.PlateCarree(), data=select_stations)

plt.savefig('C:/Users/15414/Documents/GitHub/MA Precip/MAPrecipData/Image_Products/Map_stations_with_30_year_plus_records.png')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

plt.title("Stations selected for precipitation modeling")

# Add coordiantes as axes, but do not show lines over the map itself.
grid_lines = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
grid_lines.top_labels = False
grid_lines.right_labels = False
grid_lines.xlines = False
grid_lines.ylines = False

# Set map extent and show state boundaries.
ax.set_extent([-69.75, -71.5, 41.5, 43], crs=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.STATES.with_scale('10m'), linewidth=1, facecolor='lightgrey', edgecolor='k')

# Plot stations, stations of interest, and add a legend.
ax.plot('Longitude', 'Latitude', 'o', label='Station', transform=ccrs.PlateCarree(), data=select_stations)
ax.plot('Longitude', 'Latitude', 'ko', label='Station selected \nfor modeling', transform=ccrs.PlateCarree(), data=stations_to_analyze)
ax.legend()

# Text is added to each station of interest separately since one must be offset to the right instead of the left
# COH726
plt.text(stations_to_analyze['Longitude'].iloc[0] + .04, 
         stations_to_analyze['Latitude'].iloc[0] + .04, 
         stations_to_analyze['STATION'].iloc[0], 
         bbox=dict(boxstyle="square",
                   ec=('k'),
                   fc=('w'),
                   ),
         horizontalalignment='left', transform=ccrs.PlateCarree())

# LYN614
plt.text(stations_to_analyze['Longitude'].iloc[1] + 0.04, 
         stations_to_analyze['Latitude'].iloc[1] + 0.04, 
         stations_to_analyze['STATION'].iloc[1],
         bbox=dict(boxstyle="square",
                   ec=('k'),
                   fc=('w'),
                   ),
         horizontalalignment='left', transform=ccrs.PlateCarree())

# WEY738
plt.text(stations_to_analyze['Longitude'].iloc[2] - 0.045, 
         stations_to_analyze['Latitude'].iloc[2] + 0.035, 
         stations_to_analyze['STATION'].iloc[2], 
         bbox=dict(boxstyle="square",
                   ec=('k'),
                   fc=('w'),
                   ),
         horizontalalignment='right', transform=ccrs.PlateCarree())

plt.savefig('C:/Users/15414/Documents/GitHub/MA Precip/MAPrecipData/Image_Products/Map_stations_selected_for_modeling.png')
```

### Extract and export 30-year records for selected stations

```python
def extract_30_year_record(targets):
    for station in targets:
        # Select 30 years of data from each target station.
        data = precipitation_flat[precipitation_flat["Station"] == station]
        data_30yr = data[data['Year'] > 1989]
        # Format save location string.
        save_loc = "C:/Users/15414/Documents/GitHub/MA Precip/MAPrecipData/Data_Products/{}_30yr.csv".format(station)
        # Save files.
        data_30yr.to_csv(save_loc)
```

```python
# Extract records for the stations of interest.
targets = ['COH726', 'LYN614', 'WEY738']
extract_30_year_record(targets)
```
