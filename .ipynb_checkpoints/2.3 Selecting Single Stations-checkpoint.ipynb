## 2.3 Selecting Single Stations
### Set up environment

```python
# Import packages
import cartopy
import cartopy.crs as ccrs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load station data.
unique_stations = pd.read_csv(r'C:\Users\evamb\OneDrive\Documents\Github\MAPrecipData\Data_Products\unique_stations.csv')
unique_stations.head()
```

```python
# Cartopy takes coordinates in within a ([-180°, 180°],[-90°, 90°]) space.
# Data currently requires East/West label to determine directionality (e.g. 72° could be 72°E or 72°W).
# Convert longitude values to negative numbers since they are in the western hemisphere.

unique_stations['Longitude'] = unique_stations['Longitude'] * -1
```

```python
# Load precipitation data.
precipitation_flat = pd.read_csv(r'C:\Users\evamb\OneDrive\Documents\Github\MAPrecipData\Data_Products\precipitation_flat.csv')
#precipitation_flat.head()
```

### Select and visualise stations with appropriately long records

First, I want to be working with longer records to ensure the data will capture cyclic variability. Climate is often determined over 30-year intervals, therefore I will remove stations with records that do not report data from the past 30 years. Then I will visualize the frequency of stations' monthly reporting proportions and thier locations.


```python
# Select stations whose records end in 2019 and are 30+ years long.
select_stations = unique_stations[(unique_stations['Last_year_collected'] == 2019) &
                                  (unique_stations['First_year_collected'] <= 1988)
                                 ]
```

```python
# Frequency of monthly reporting proportions e.g. How often did a station report data during its operation? 

plt.hist(select_stations['Percent_months_reported'], bins=20)
plt.title("Frequency of monthly reporting proportions\n by stations with 30+ year records")
plt.xlabel("Proportion of months reported")
plt.ylabel("Frequency")
plt.savefig(r'C:\Users\evamb\OneDrive\Documents\Github\MAPrecipData\Data_Products\Frequency-monthly-reporting-proportion.png')
```

```python
# Map of stations' locations color-coded by proportion of months reported.
ax = plt.axes(projection=ccrs.PlateCarree())

plt.title("Proportion of months reported\nby stations with 30+ year records")

# Set map extent and show state boundaries.
ax.set_extent([-69.5, -74, 41, 43.5], crs=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.STATES.with_scale('10m'), linewidth=1, edgecolor='k',)

# Add coordiantes as axes, but do not show lines over the map itself.
grid_lines = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
grid_lines.top_labels = False
grid_lines.right_labels = False
grid_lines.xlines = False
grid_lines.ylines = False

# Plot stations, stations of interest, and add a legend.
clrbar = ax.scatter(x='Longitude', y='Latitude', c='Percent_months_reported', data=select_stations, transform=ccrs.PlateCarree())
plt.colorbar(clrbar, ax=ax, shrink=0.6)

# Save figure.
plt.savefig(r'C:\Users\evamb\OneDrive\Documents\Github\MAPrecipData\Image_Products\Map-monthly-reporting-proportion.png')
```

### Select and visualise stations in the Boston metro area

I identified three stations from the area surrounding Boston, those in Cohasset (station COH726), Lynn (station LYN614), and Weymouth (station WEY738). While the first two stations report >90% coverage, Station WEY738 reports 78% coverage. I am specifically including it to examine the effect of record completeness on the models generated. 

```python
# Select three stations around Boston metro area. 
stations_to_analyze = select_stations[(select_stations['STATION'] == 'COH726') |
                                      (select_stations['STATION'] == 'LYN614') |
                                      (select_stations['STATION'] == 'WEY738')
                                     ]
stations_to_analyze.head()
```

```python
# Magnified map of stations' locations with Cohasset, Lynn, and Weymouth stations labeled.
ax = plt.axes(projection=ccrs.PlateCarree())

plt.title("Stations selected for precipitation modeling")

# Add coordiantes as axes, but do not show lines over the map itself.
grid_lines = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
grid_lines.top_labels = False
grid_lines.right_labels = False
grid_lines.xlines = False
grid_lines.ylines = False

# Set map extent and show state boundaries.
ax.set_extent([-69.75, -71.5, 41.5, 43], crs=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.STATES.with_scale('10m'), linewidth=1, edgecolor='k')

# Plot stations, stations of interest, and add a legend.
ax.plot('Longitude', 'Latitude', 'o', label='Station', transform=ccrs.PlateCarree(), data=select_stations)
ax.plot('Longitude', 'Latitude', 'ko', label='Station selected \nfor modeling', transform=ccrs.PlateCarree(), data=stations_to_analyze)
ax.legend()

# Text is added to each station of interest separately since one must be offset to the right instead of the left
# COH726
plt.text(stations_to_analyze['Longitude'].iloc[0] + .04, 
         stations_to_analyze['Latitude'].iloc[0] + .04, 
         stations_to_analyze['STATION'].iloc[0], 
         bbox=dict(boxstyle="square",
                   ec=('k'),
                   fc=('w'),
                   ),
         horizontalalignment='left', transform=ccrs.PlateCarree())

# LYN614
plt.text(stations_to_analyze['Longitude'].iloc[1] + 0.04, 
         stations_to_analyze['Latitude'].iloc[1] + 0.04, 
         stations_to_analyze['STATION'].iloc[1],
         bbox=dict(boxstyle="square",
                   ec=('k'),
                   fc=('w'),
                   ),
         horizontalalignment='left', transform=ccrs.PlateCarree())

# WEY738
plt.text(stations_to_analyze['Longitude'].iloc[2] - 0.045, 
         stations_to_analyze['Latitude'].iloc[2] + 0.035, 
         stations_to_analyze['STATION'].iloc[2], 
         bbox=dict(boxstyle="square",
                   ec=('k'),
                   fc=('w'),
                   ),
         horizontalalignment='right', transform=ccrs.PlateCarree())

plt.savefig(r'C:\Users\evamb\OneDrive\Documents\Github\MAPrecipData\Image_Products\Map-selected-stations.png')
```

### Extract and export 30-year records for selected stations

```python
def extract_30_year_record(targets):
    for station in targets:
        # Select 30 years of data from each target station.
        data = precipitation_flat[precipitation_flat["Station"] == station]
        data_30yr = data[data['Year'] > 1989]
        # Format save location string.
        save_loc = "C:/Users/evamb/OneDrive/Documents/Github/MAPrecipData/Data_Products/{}_30yr.csv".format(station)
        # Save files.
        data_30yr.to_csv(save_loc)
        
```

```python
# Extract records for the stations of interest.
station_names = ['COH726', 'LYN614', 'WEY738']
extract_30_year_record(station_names)
```
