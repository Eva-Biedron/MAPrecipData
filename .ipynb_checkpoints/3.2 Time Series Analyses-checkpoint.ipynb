### Seasonal ARIMA models

#### Model choice and parameterization

I utilize the statsmodels package (v0.14.0) to perform seasonal and predictive modeling (https://www.statsmodels.org/stable/index.html). I apply a Seasonal-Trend decomposition using LOESS (STL) to separate and visualize patterns of seasonality and trend in the precipitation data. I apply a Autoregressive Integrated Moving-Average (ARIMA) model to produce predictions fore each station. Statsmodels’ STLForecast function integrates an STL analysis with an ARIMA model in a one-step function.

#### Seasonal and Trend decomposition models using Loess (STL)

STL models separate the data into three components: seasonality, trend, and the remaining variation in the data. I chose a periodicity of 12 months to reflect yearly precipitation patterns. Other periodic global climate teleconnections like Atlantic multidecadal oscillation (60 to 85 year periodicity) or El Niño-southern oscillation (2 to 7 year periodicity) appear to affect precipitation patterns (as measured by baseflow) in New England to varying degrees (Chen et al. 2021). However, the periodicity of these teleconnections is not constant and cannot be modeled using the implementation of seasonal-trend decomposition. The impacts of these teleconnections cannot be separated at this level of analysis. Robustness was set to True since data is non-normal (Cleveland et al. 1990, p.14). I let the trend smoother be calculated automatically.

#### ARIMA Model Parameterization

ARIMA (Autoregressive Integrated Moving-Average) models assume stationary data. Non-stationarity can be detected by a pattern of slow decay in autocorrelation function (ACF) plots. Despite a few non-sequential but significant lags, none of the stations show this characteristic decay in ACF values. Therefore, the data do not show a need to perform additional differencing and integration (d) is set to 0 for all models.

Including Ar or MA terms with order >13 (one year of lag) is unappealing since one term is calculated for each order of each term. This means a model could become overfit very quickly if too many terms are added. Although all ACF and PACF plots show significant lags greater than 13, I will not test these models as their significance may be related to earlier autocorrelations (e.g. a one year autocorrelation may 'echo' at two years). I will model different combinations of parameters for each station, but I am setting an upper limit on p and q values at 13.

Significant lags in the partial autocorrelation function (PACF) plot at all stations indicate the need for additional autoregressive components; additionally, near significant lags can be investigated as well (Hyndmann & Athanasopoulos, 2018). Station COH726 shows significant lags (\*\*) and nearly significant lags (\*) at 3\*\*, 13\*, 16\*, 17\*\*, 22\*. Station LYN614 shows lags at 3\*\*, 12\*, 14\*\*, 16\*\*. Station WEY736 shows lags at 1\*\*, 2\*, 3\*\*, 5\*\*, 6\*, 9\*\*, 12\*, 24\*\*.

Similary, significant lags in the autocorrelation function (ACF) plot at all stations indicate the need for additional moving average components (Hyndmann & Athanasopoulos, 2018). Station COH726 shows significant lags (\*\*) and nearly significant lags (\*) at 3\*\*, 6\*, 9\*, 12\*, 13\*, 16\*\*, 17\*\*, 22\*. Station LYN614 shows lags at 3\*\*, 12\*, 14\*\*, 16\*, 20\*. Station WEY738 shows lags at 1\*\*, 2\*, 3\*\*, 5\*, 7\*\*, 9\*\*, 12\*\*, 21\*, 24\*\*, 26\*.

For each station, I will create models based on the following p, d, and q parameter combinations. 

| Station | p | d | q |
| :-: | :-: | :-: | :-: |
| COH726 | 0, 3, 13 | 0 | 0, 3, 6, 9, 12 |
| LYN614 | 0, 3, 12 | 0 | 0, 3, 12 |
| WEY738 | 0, 1, 2, 3, 5, 6, 12 | 0 | 0, 1, 2, 3, 5, 7, 9, 12 |

#### Model Comparison

I will use Akaike's Information Criterion to compare the amount of information gained by adding increasing model terms. The best fitting model for each station is the one with the lowest AIC score. AIC scores cannot be compared across datasets, only between models created from one parent dataset. Comparing AIC (Akaike's Information Criterion) values is also an established method to optimize p and q values for modeling a distribution (Hyndman & Athanasopoulos, 2018), so I feel comfortable using it to evaluate the effects of different model parameterizations.

###  Format parameters

```python
### Create function to format parameters.

# Create a function to format parameter dictionaries for use in Sarimax modeling.
def format_param_dicts(sarimax_params):
    """
    This function's purpose is to format parameter dictionaries for use in future Sarimax modeling.
    
    arguments: sarimax params, a list of the station name, a list of p values, a list of d values, and a list of q values.
    returns: param_dict, a dictionary of model trial names (keys, strings) and parameters (values, a tuple)
    
    """
    
    # Separate lists.
    p_s = sarimax_params[1]
    d_s = sarimax_params[2]
    q_s = sarimax_params[3]
    prefix = sarimax_params[0]
    
    # Create a dictionary to hold the results.
    param_dict = {}
    counter = 1
    
    # Iterate over each set of p, d, and q values.
    for a in p_s:
        for b in d_s:
            for c in q_s:
                # Create a tuple from the parameters.
                param_combo = (a, b, c)
                # Create a name for this parameter set.
                name = prefix + str("-") + str(counter)
                # Iterate to next trial number.
                counter = counter + 1
                # Add the name and parameters to the dictionary.
                param_dict.update({name: param_combo})
    
    # Output results.
    return param_dict
                
```

```python
### Format parameters.

# Set up input for format_param_dicts function.
COH726_sarimax_params = ['COH726', [ 0, 3, 13], [0], [0, 3, 6, 9, 12]]
LYN614_sarimax_params = ['LYN614', [0, 3, 12], [0], [0, 3, 12]]
WEY738_sarimax_params = ['WEY738', [0, 1, 2, 3, 5, 6, 12], [0], [0, 1, 2, 3, 5, 7, 9, 12]]

# Format parameters for all stations.
COH_param_dict = format_param_dicts(COH726_sarimax_params)
LYN_param_dict = format_param_dicts(LYN614_sarimax_params)
WEY_param_dict = format_param_dicts(WEY738_sarimax_params)

# Combine all stations' parameters into one dictionary.
all_station_all_sarimax_parameters = {**COH_param_dict, **LYN_param_dict, **WEY_param_dict}

# Convert to dataframe.
sarimax_parameter_combos = pd.DataFrame(all_station_all_sarimax_parameters)

# Transpose dataframe to show trial name as index column.
sarimax_parameter_combos = sarimax_parameter_combos.transpose()

# Set index column name.
sarimax_parameter_combos.index.name = 'Station Trial Name'

# Set columns to parameter names.
sarimax_parameter_combos.columns = ['p', 'd', 'q']

# Copy table to hold results of models on normalized data.
sarimax_parameter_combos_norm = sarimax_parameter_combos

# Format dictionaries of models and parameters.
COH_param_dict = format_param_dicts(COH726_sarimax_params)
LYN_param_dict = format_param_dicts(LYN614_sarimax_params)
WEY_param_dict = format_param_dicts(WEY738_sarimax_params)
```

### Run models on normalized & non-normalized data & calculate AIC values

#### Non-normalized data

```python
### Create function to run SARIMA models.

# Create function to run SARIMA models, calculate AIC values, and return a table of AIC results.
def SARIMA_table_AIC(station, parameter_dict):
    """
    This functions purpose is perform a joint STL analysis and ARIMA model on a 
    data set and return the input parameters and resulting AIC value.
   
    Arguements: 
    station- A dataset indexed by month
    parameter_dict- a dictionary of model names (keys) and parameters (values).
    
    Returns: A list of lists of parameter values and AIC results.
    """
    
    months = 12 # n_p (period) is monthly.
    seasons = int(months + months/2 + 1) # n_s (seasonal smoother) 
    
    # Create lists to hold values of interest.
    param_values = []
    aic_values = []
    forecasts = []
    
    # Iterate over each set of parameters when making models.
    for param in parameter_dict.values():
        # This statement indicates the next parameter set is being modeled/throwing warnings.
        print(param)
        
        # Fit model.
        stlf = STLForecast(station, ARIMA, model_kwargs=dict(order=param, trend="t"), period = months, seasonal = seasons, robust = True)
        stlf_res = stlf.fit()
        
        # Forecast 4 years into the future.
        forecast = stlf_res.forecast(48)
        
        # Get the aic (Akaike's Information Criterion)
        res_html = stlf_res.summary().tables[0].as_html()
        res_df = pd.read_html(res_html)
        aic = res_df[0][3][2]
        
        param_str = str(param)
        param_values.append(param_str)
        
        aic_values.append(aic)
        
        forecasts.append(forecast)
        
    return param_values, aic_values, forecasts
```

```python
### Calculate and save AIC values.

# Calculate AIC values for all models in the dictionary.
COH_AICs = SARIMA_table_AIC(COH726, COH_param_dict)

# Repeat for LYN614.
LYN_AICs = SARIMA_table_AIC(LYN614, LYN_param_dict)

# Repeat for WEY738.
WEY_AICs = SARIMA_table_AIC(WEY738, WEY_param_dict)

# Add all AIC values to a list of lists.
all_AICs = []
all_AICs.append(COH_AICs[1])
all_AICs.append(LYN_AICs[1])
all_AICs.append(WEY_AICs[1])

# Flatten the list of lists.
AICs = [x for station_AICs in all_AICs for x in station_AICs]

# Add the flattened list as a new column in the model and parameter table.
sarimax_parameter_combos['AIC value'] = AICs

# Save the table.
sarimax_parameter_combos.to_csv('C:/Users/evamb/OneDrive/Documents/Github/MAPrecipData/Data_Products/params_AIC_actual.csv', index = True)
sarimax_parameter_combos
```

```python
### Rerun best models.

# Set up variables for rerunning best models. These are used by the models for all three stations.
months = 12 # n_p (period) is monthly.
seasons = int(months + months/2 + 1) # n_s (seasonal smoother) 

# Rerun model with lowest AIC score for COH726.
COH_stlf = STLForecast(COH726, ARIMA, model_kwargs=dict(order=(3, 0, 3), trend="t"), period = months, seasonal = seasons, robust = True)
COH_stlf_res = COH_stlf.fit()

# Output  summary of STL Decomposition and SARIMAX Results 
#COH_stlf_res.summary()

# Store the model.
COH_forecast = COH_stlf_res.model_result

# Forecast values for modeled data range.
COH_predictions = COH_forecast.get_prediction(0)

# Add the confidence intervals for each month's prediction to a dataframe.
COH_validation_predictions = pd.DataFrame(COH_predictions.conf_int(alpha=0.05))

# Add  the predicted mean to the dataframe.
COH_validation_predictions["mean"] = COH_predictions.predicted_mean

# Add the actual values to the table for easy comparison.
COH_validation_predictions["actual"] = COH726["Precipitation"]

# Predict values beyond modeled data range for 3 years and add to a new dataframe.
COH_future_pred = pd.DataFrame(COH_forecast.forecast(36))

# Add the predicted values to the dataframe with forcasted and actual values
COH_validation_predictions = pd.concat([COH_validation_predictions, COH_future_pred], axis=0)

# Reset index to make a column with the month-year data.
COH_validation_predictions = COH_validation_predictions.reset_index()

# Rename the new column.
COH_validation_predictions = COH_validation_predictions.rename(columns={"index":"Date"})

# Ensure there are no duplicates between the forecasted and predicted values.
COH_validation_predictions = COH_validation_predictions.drop_duplicates(keep = 'first')

# Repeat for LYN614.
LYN_stlf = STLForecast(LYN614, ARIMA, model_kwargs=dict(order=(3, 0, 3), trend="t"), period = months, seasonal = seasons, robust = True)
LYN_stlf_res = LYN_stlf.fit()
LYN_forecast = LYN_stlf_res.model_result

# Get forecast.
LYN_predictions = LYN_forecast.get_prediction(0)

# Add forecast and actual data to new dataframe.
LYN_validation_predictions = pd.DataFrame(LYN_predictions.conf_int(alpha=0.05))
LYN_validation_predictions["mean"] = LYN_predictions.predicted_mean
LYN_validation_predictions["actual"] = LYN614["Precipitation"]

# Get predictied mean.
LYN_future_pred = pd.DataFrame(LYN_forecast.forecast(36))

# Add predicted mean to dataframe and clean up.
LYN_validation_predictions = pd.concat([LYN_validation_predictions, LYN_future_pred], axis=0)
LYN_validation_predictions = LYN_validation_predictions.reset_index()
LYN_validation_predictions = LYN_validation_predictions.rename(columns={"index":"Date"})
LYN_validation_predictions = LYN_validation_predictions.drop_duplicates(keep = 'first')
#LYN_validation_predictions

# Repeat for WEY738.
WEY_stlf = STLForecast(WEY738, ARIMA, model_kwargs=dict(order=(12, 0, 0), trend="t"), period = months, seasonal = seasons, robust = True)
WEY_stlf_res = WEY_stlf.fit()
WEY_forecast = WEY_stlf_res.model_result

# Get forecast.
WEY_predictions = WEY_forecast.get_prediction(0)

# Add forecast and actual data to new dataframe.
WEY_validation_predictions = pd.DataFrame(WEY_predictions.conf_int(alpha=0.05))
WEY_validation_predictions["mean"] = WEY_predictions.predicted_mean
WEY_validation_predictions["actual"] = WEY738["Precipitation"]

# Get predictied mean.
WEY_future_pred = pd.DataFrame(WEY_forecast.forecast(36))

# Add predicted mean to dataframe and clean up.
WEY_validation_predictions = pd.concat([WEY_validation_predictions, WEY_future_pred], axis=0)
WEY_validation_predictions = WEY_validation_predictions.reset_index()
WEY_validation_predictions = WEY_validation_predictions.rename(columns={"index":"Date"})
WEY_validation_predictions = WEY_validation_predictions.drop_duplicates(keep = 'first')
#WEY_validation_predictions
```
