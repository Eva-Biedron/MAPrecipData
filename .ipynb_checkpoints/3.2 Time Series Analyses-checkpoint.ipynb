### SARIMAX model and parameters

A SARIMA(X) model is a variant of an ARIMA model that can account for seasonal cycles in data.

S   Seasonal
Ar  Autoregressive
I   Integrated
Ma  Moving Average
(X   w/ eXogenous features)

I will not be supplementing with Exogenous features (i.e., additional climate data), although the model I use does allow for the addition of other predicive data.

#### Stationarity and Seasonality

While there are a few significant lags, none of the stations show a pattern of slow decay in thier autocorrelation function (ACF) values. This does not indicate non-stationarity. The  model expects stationary data (aside from the seasonal component) so we do not violate that assumption.

Seasonality is expected in monthly climate data. The most obvious cycle is yearly with a period of 12 months. The seasonal component of the SARIMAX model will be calculated by the Seasonal Trend Decomposition with Loess (STL), below).

#### Moving Average Order
The ACF plots show significant spikes at the following lags:
COH: 3, 6, 17
LYN: NA
WEY: 7 
However, it is important to note that none of the lags within stations are much greater than the others and none of the autocorrelation values of the lags are greater than 0.2. This appears to be more indicative of a white noise process than a higher order moving average model. This indicates q should be set to either 0 for all three models or 3, 0, and 7, respectively. 

#### Autoregressive Order
The ACF plots do not show an exponentially decreasing pattern, indicating a higher order autoregressive component (than 1) is requred to model these records. The partial autocorrelation plots indicate significant lags at:
COH: 3, 6, 17
LYN: 3 
WEY: 7 
Again, several of these values are barely significant and are not highly autocorrelated. Therefore, p should should be set to either 0 or 1 for all three models or 6, 3, and 7, respectively.

#### Model Overfitting
Including Ar or MA terms with order >3 is unappealing since one term is calculated for each order of each term. This means a model could become overfit very quickly if too many terms are added. Although the ACF and PACF plots show significant lags at 17 and 17 for the Cohasset station, I don't want to fit a model with 34+ terms

#### Integration
The 1 month ACF lag is not significantly different from 0 at all three stations, indicating another degree of differencing is not required (https://people.duke.edu/~rnau/411arim2.htm). Therefore, d = 0 for all stations.

#### Comparing models

The best fitting model for each station is the one with the lowest AIC (Akaike's Information Criterion) value. AIC scores cannot be compared across datasets, only between models created from one parent dataset.

Comparing AIC (Akaike's Information Criterion) values is also an established method to determine p and q values for modeling a distribution. 

#### Model Parameter Combinations
COH: p=0/3/6, d=0, q=0/3
LYN: p=0, d=0, q=0
WEY: p=0/7, d=0, q=0/7

### Seasonal Trend decompositon with Loess (STL)

See Clevelandet al. (1990) STL: A Seasonal-Trend Decomposition Procedure Based on LOESS. Journal of Official Statistics, 6. 

See also: 
- 

Robustness was set to True since data is non-normal (Cleveland et al. 1990, p.14). Let n_t (trend smoother) be calculated automatically.

```python
# Set parameters.
months = 12 # n_p (period) is monthly.
seasons = int(months + months/2 + 1) # n_s (seasonal smoother) 
```

```python
def save_STL_plot(station_info):
    """
    
    """
    # Unpack input.
    target_stations = station_info[0]
    station_names = station_info[1]
    
    counter = 0
    
    for station in target_stations:
    
        res = STL(station, period = months, seasonal = seasons, robust = True).fit()
        fig = res.plot()
        fig.axes[0].set_ylabel('Observed')
        plt.xlabel('Year')
      
        # Format save location string.
        save_loc = 'C:/Users/evamb/Documents/Github/MAPrecipData/Image_Products/ACF_{}.png'.format(station_names[counter])
        # Save files.
        plt.savefig(save_loc)
    
        # Increase counter to use the name of the next station in the next loop.
        counter = counter + 1
```

```python
# Apply function to create and save autocorrelation plots.
station_info = [[COH726, LYN614, WEY738], ['COH726', 'LYN614', 'WEY738']]
save_STL_plot(station_info)
```

```python
def SARIMAX_table_AIC(station, parameter_dict):
    months = 12 # n_p (period) is monthly.
    seasons = int(months + months/2 + 1) # n_s (seasonal smoother) 
    
    param_values = []
    aic_values = []
    
    for param in parameter_dict.values():
        stlf = STLForecast(station, ARIMA, model_kwargs=dict(order=param, trend="t"), period = months, seasonal = seasons, robust = True)
        stlf_res = stlf.fit()
        
        res_html = stlf_res.summary().tables[0].as_html()
        res_df = pd.read_html(res_html)
        aic = res_df[0][3][2]
        
        param_str = str(param)
        param_values.append(param_str)
        aic_values.append(aic)
        
    results = pd.DataFrame([param_values, aic_values])

    return param_values, aic_values
```

```python
COH_param_dict = {
    "COH_000": (0,0,0),
    "COH_003": (0,0,3),
    "COH_303": (3,0,3),
    "COH_603": (6,0,3),
    "COH_300": (3,0,0),
    "COH_600": (6,0,0)
}

COH_AICs = SARIMAX_table_AIC(COH726, COH_param_dict)
```
