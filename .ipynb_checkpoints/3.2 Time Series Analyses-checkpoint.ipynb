```python
# Import packages
#import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import STL

""" not using
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
from scipy import stats

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.seasonal import STL

import copy
import datetime
import math"""
```

```python
# Load 30-year records from each target station.
COH726 = pd.read_csv(r'C:/Users/evamb/Documents/Github/MAPrecipData/Data_Products/COH726_30yr.csv')
LYN614 = pd.read_csv(r'C:/Users/evamb/Documents/Github/MAPrecipData/Data_Products/LYN614_30yr.csv')
WEY738 = pd.read_csv(r'C:/Users/evamb/Documents/Github/MAPrecipData/Data_Products/WEY738_30yr.csv')

# Test: The records are 30 years long.
n = 360 # Length of record in months.
assert (len(COH726) == n) & (len(LYN614) == n) & (len(WEY738) == n), "One or more records do not contain exactly 30 years of data." 

# Test: Record lengths are equal across datasets.
assert (len(COH726) == len(LYN614)) & (len(WEY738) == len(LYN614)), "One or more records are unequal lengths."
```

```python
# Index all records by date.

dates = pd.date_range("1-1-1990", periods=len(COH726), freq="M")
dates = pd.to_datetime(dates)

COH726["Date"] = dates
COH726 = COH726.set_index("Date")
COH726 = COH726[['Precipitation']]
COH726 = COH726.dropna()

LYN614["Date"] = dates
LYN614 = LYN614.set_index("Date")
LYN614 = LYN614[['Precipitation']]
LYN614 = LYN614.dropna()

WEY738["Date"] = dates
WEY738 = WEY738.set_index("Date")
WEY738 = WEY738[['Precipitation']]
WEY738 = WEY738.dropna()
```

```python
fig, axes = plt.subplots(3, 1, figsize = (6,8))
plot_acf(COH726, alpha=0.05, zero=False, auto_ylims=True, title='COH726', ax=axes[0])
plot_acf(LYN614, alpha=0.05, zero=False, auto_ylims=True, title='LYN614', ax=axes[1])
plot_acf(WEY738, alpha=0.05, zero=False, auto_ylims=True, title='WEY738', ax=axes[2])
fig.tight_layout()
axes[2].set(xlabel='Lags (months)') 
fig.suptitle('Autocorrelation in monthly precipitation records', y=1.02, x = 0.54, fontsize=14)
plt.savefig('C:/Users/evamb/Documents/Github/MAPrecipData/Image_Products/ACF_station_precipitation.png')
```

```python
months = 12
seasons = int(months + months/2 + 1)

res = STL(COH726, period = months, seasonal = seasons).fit()
fig = res.plot()
fig.axes[0].set_ylabel('Observed')
plt.xlabel('Year')
plt.show()
```

```python
res = STL(LYN614, period = months, seasonal = seasons).fit()
fig = res.plot()
fig.axes[0].set_ylabel('Observed')
plt.xlabel('Year')
plt.show()
```

```python
res = STL(WEY738, period = months, seasonal = seasons).fit()
fig = res.plot()
fig.axes[0].set_ylabel('Observed')
plt.xlabel('Year')
plt.show()
```
